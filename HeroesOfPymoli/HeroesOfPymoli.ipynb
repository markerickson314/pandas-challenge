```python
# import dependencies
import pandas as pd

# path to file
csv_path = "Resources/purchase_data.csv"

# read csv and create dataframe
purchase_data = pd.read_csv(csv_path)
```

```python
# inspect data
purchase_data
```

```python
# Display total number of players in a dataframe
player_count = purchase_data["SN"].nunique()
total_players_df = pd.DataFrame({"Total Players": [player_count]})
total_players_df
```

```python
# create variables and put into dataframe
unique_items = purchase_data["Item ID"].nunique()
average_price = "${:.2f}".format(purchase_data["Price"].mean())
purchases = purchase_data["Purchase ID"].count()
total_revenue = "${:,.2f}".format(purchase_data["Price"].sum())
purchasing_analysis_df = pd.DataFrame({ "Number of Unique Items": [unique_items],
                                        "Average Price": average_price,
                                        "Number of Purchases": purchases,
                                        "Total Revenue": total_revenue})
purchasing_analysis_df
```

```python
# remove duplicate players
unique_df = purchase_data.drop_duplicates(subset="SN", keep="first")

# calculate number of players by gender, get sum
males = len(unique_df[unique_df["Gender"] == "Male"])
females = len(unique_df[unique_df["Gender"] == "Female"])
others = len(unique_df[unique_df["Gender"] == "Other / Non-Disclosed"])
total = males + females + others
percent_males = "{:.2%}".format(males/total) 
percent_females = "{:.2%}".format(females/total)
percent_others = "{:.2%}".format(others/total)

# create summary data frame
gender_demo_df = pd.DataFrame({"Total Count": [males, females, others],
                               "Percentage of Players": [percent_males, percent_females, percent_others]},
                               index = ["Male", "Female", "Other / Non-Disclosed"])
gender_demo_df
```

```python
# groupy gender
grouped_gender_df = purchase_data.groupby(["Gender"])

# calculate columns
purchase_count = grouped_gender_df["Purchase ID"].count()
average_purchase_price = grouped_gender_df["Price"].mean()
total_purchase_price = grouped_gender_df["Price"].sum()

# create list of per person by gender
per_person = [81, 484, 11]
average_per_person = total_purchase_price / per_person

# create summary dataframe
gender_analysis_df = pd.DataFrame({"Purchase Count": purchase_count,
                                   "Average Purchase Price": average_purchase_price,
                                   "Total Purchase Value": total_purchase_price,
                                   "Avg Total Purchase per Person": average_per_person    
})
gender_analysis_df
```
