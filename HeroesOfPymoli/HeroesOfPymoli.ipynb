```python
# import dependencies
import pandas as pd

# path to file
csv_path = "Resources/purchase_data.csv"

# read csv and create dataframe
purchase_data = pd.read_csv(csv_path)
```

```python
# inspect data
purchase_data
```

```python
# Display total number of players in a dataframe
player_count = purchase_data["SN"].nunique()
total_players_df = pd.DataFrame({"Total Players": [player_count]})
total_players_df
```

```python
# create variables and put into dataframe
unique_items = purchase_data["Item ID"].nunique()
average_price = "${:.2f}".format(purchase_data["Price"].mean())
purchases = purchase_data["Purchase ID"].count()
total_revenue = "${:,.2f}".format(purchase_data["Price"].sum())
purchasing_analysis_df = pd.DataFrame({ "Number of Unique Items": [unique_items],
                                        "Average Price": average_price,
                                        "Number of Purchases": purchases,
                                        "Total Revenue": total_revenue})
purchasing_analysis_df
```

```python
# remove duplicate players
unique_df = purchase_data.drop_duplicates(subset="SN", keep="first")

# calculate number of players by gender, get sum
males = len(unique_df[unique_df["Gender"] == "Male"])
females = len(unique_df[unique_df["Gender"] == "Female"])
others = len(unique_df[unique_df["Gender"] == "Other / Non-Disclosed"])
total = males + females + others
percent_males = "{:.2%}".format(males/total) 
percent_females = "{:.2%}".format(females/total)
percent_others = "{:.2%}".format(others/total)

# create summary data frame
gender_demo_df = pd.DataFrame({"Total Count": [males, females, others],
                               "Percentage of Players": [percent_males, percent_females, percent_others]},
                               index = ["Male", "Female", "Other / Non-Disclosed"])
gender_demo_df
```

```python
# groupy gender
grouped_gender_df = purchase_data.groupby(["Gender"])

# calculate columns
purchase_count = grouped_gender_df["Purchase ID"].count()
average_purchase_price = round(grouped_gender_df["Price"].mean(), 2)
total_purchase_price = grouped_gender_df["Price"].sum()

# create list of per person by gender
per_person = [81, 484, 11]
average_per_person = round(total_purchase_price / per_person, 2)

# create summary dataframe
gender_analysis_df = pd.DataFrame({"Purchase Count": purchase_count,
                                   "Average Purchase Price": average_purchase_price,
                                   "Total Purchase Value": total_purchase_price,
                                   "Avg Total Purchase per Person": average_per_person    
                                    })
gender_analysis_df
```

```python
# create bins <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40+
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# create names for the bins
age_ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# new column "Age Ranges" on unique_df
unique_df["Age Ranges"] = pd.cut(unique_df["Age"], bins, labels=age_ranges, include_lowest=True)
```

```python
# Groupby age ranges
age_ranges_df = unique_df.groupby("Age Ranges")

# calculate columns
total_count = age_ranges_df["Purchase ID"].count()
percent_count = round(total_count / 576 * 100, 2)

# create summary dataframe
age_demographics_df = pd.DataFrame ({"Total Count": total_count,
                                     "Percentage of Players": percent_count
                                    })
age_demographics_df
```

```python
# create bins <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40+
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# create names for the bins
age_ranges = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# new column "Age Ranges" on purchase_data
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=age_ranges, include_lowest=True)
```

```python
# Groupby age ranges
age_purchasing_df = purchase_data.groupby("Age Ranges")

# calculate columns
purchase_count_1 = age_purchasing_df["Purchase ID"].count()
average_purchase_price_1 = round(age_purchasing_df["Price"].mean(), 2)
total_purchase_price_1= age_purchasing_df["Price"].sum()

# create list of per person by age range
per_person_1 = [17, 22, 107, 258, 77, 52, 31, 12]
average_per_person_1 = round(total_purchase_price_1 / per_person_1, 2)
average_per_person_1

# create summary dataframe
age_analysis_df = pd.DataFrame({"Purchase Count": purchase_count_1,
                                   "Average Purchase Price": average_purchase_price_1,
                                   "Total Purchase Value": total_purchase_price_1,
                                   "Avg Total Purchase per Person": average_per_person_1    
                                    })
age_analysis_df
```

```python
# groupby SN
grouped_SN_df = purchase_data.groupby(["SN"])

# calculate columns
purchase_count_2 = grouped_SN_df["Purchase ID"].count()
average_purchase_price_2 = round(grouped_SN_df["Price"].mean(), 2)
total_purchase_price_2 = grouped_SN_df["Price"].sum()

# create summary dataframe
top_spenders_df = pd.DataFrame({"Purchase Count": purchase_count_2,
                                   "Average Purchase Price": average_purchase_price_2,
                                   "Total Purchase Value": total_purchase_price_2,   
                                    })

top_spenders_sorted_df = top_spenders_df.sort_values("Total Purchase Value", ascending=False)
top_spenders_sorted_df.head()
```
